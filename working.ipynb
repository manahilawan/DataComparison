# Combining Datasets

```python
pak['Country']= 'Pakistan'
turkiye['Country']= 'Turkiye'
```

```python
turkiye.head(5)
```

```python
combined = pd.concat([pak, turkiye], ignore_index=True)
```

```python
combined[['Country','Value']]
combined.head(3)
# combined.tail(3)
```

```python
combined[['Country','Year', 'Value']].sample(10)
```

```python
pak_head= combined[combined['Country']=='Pakistan'].head(3)
```

```python
tur_head= combined[combined['Country']=='Turkiye'].head(3)
```

```python
small= pd.concat([pak_head, tur_head])
small
```

```python
small[['Country','Year', 'Value']]
```
